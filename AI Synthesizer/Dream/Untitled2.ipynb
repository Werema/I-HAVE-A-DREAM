```python
from textblob import TextBlob
import csv

file_path =r"C:\Users\Avom\Downloads\MartinLutherMalcolmX.txt"
sentiment_csv_path =r"C:\Users\Avom\Downloads\DreamSentiment.csv"

fieldnames = ['SENTENCE_ID','POLARITY','SUBJECTIVEITY', 'SENTENCE','STRONG OPINION']

sentence_ID = 0
with open(sentiment_csv_path, 'w') as the_csv:
    writer = csv.DictWriter(the_csv, fieldnames=fieldnames)
    writer.writeheader()
    the_csv.close()
    
file = open(file_path, 'r',encoding='utf-8')
open_file = TextBlob(file.read())

file_sentences = open_file.sentences

print(file_sentences)

for sentence in file_sentences:
    f = sentence.sentiment.polarity
    f2 = sentence.sentiment.subjectivity
    
    if abs(f) >= 0.7 and f2 >= 0.7:
        h = 1
    else:
        h = 0
    with open(sentiment_csv_path, 'a') as the_csv:
        writer = csv.DictWriter(the_csv, fieldnames = fieldnames)
        writer.writerow({fieldnames[0]:sentence_ID, fieldnames[1]:f,fieldnames[2]:f2,fieldnames[3]:sentence, fieldnames[4]:h})
        the_csv.close()
        
        sentence_ID += 1


    
    
```
